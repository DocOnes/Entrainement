```python
from random import randint
import time

#Création de la liste à trier
longueurliste=int(input("combien de valeur dans la liste ?:"))
liste=[randint(0,100000) for i in range(longueurliste)]


def tri(liste):#je parcours tout les nombre de 0 au maximum de la liste et quand il y en a un dans la liste je le met à sa place dans la liste
    a=0
    for i in range(max(liste)+1):
        if i in liste:
            liste[liste.index(i)],liste[a]=liste[a],i
            a+=1

def tri2(liste): #tri bulle
    for i in range(len(liste)-1):
        for j in range(len(liste)-(i+1)):
            if liste[j]>liste[j+1]:
                liste[j],liste[j+1]=liste[j+1],liste[j]

def tri3(liste):#je créer une autre liste vide avec le nombre max de case donc si le max c'est 100 il y aura une liste vide de 100 case
    listsort=[]
    for i in range(max(liste)+1):   #après i prend les valeurs de la liste de base et les met à leur place c'est a dire que il prend 5 et le met dans la case vide 5
        listsort.append([])
    [listsort[i].append(i) for i in liste]
    listsort = [i for i in listsort if i != []]     #après j'enleve les cases qui sont vides
    print(listsort)

print("liste de base :",liste)
début = time.time()

tri(liste)

fin = time.time()
#print("liste trier :",liste)



temps_execution = fin-début
print("Temps d'execution de :",temps_execution,"secondes")
```
